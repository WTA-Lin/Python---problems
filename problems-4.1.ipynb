## ⭐ Úloha 05
Prečítajte súbor CSV a nájdite všetky položky, ktorých predajná cena je viac ako dvojnásobok nákupnej ceny. Vytlačte názvy položiek a percento prirážky.

```python
df = pd.read_csv("csvsample.csv")
price_sale = df[df["Type"] == "Sale"] 
price_purchase = df[df["Type"] == "Purchase"]    
```

```python
merged = pd.merge(price_sale, price_purchase, on="Item", suffixes=("_sale", "_purchase"))
spojene_polozky = merged[["Item", "Unit Price ($)_sale", "Unit Price ($)_purchase"]]
print(spojene_polozky)
```

```python
filtered = merged[merged["Unit Price ($)_sale"] >= 2 * merged["Unit Price ($)_purchase"]]
percento_prirazky = ((filtered["Unit Price ($)_sale"] - filtered["Unit Price ($)_purchase"]) / filtered["Unit Price ($)_purchase"]) * 100

print(filtered[["Item"]].assign(Percento_prirazky=percento_prirazky))
```

## ⭐ Úloha 09
Prečítajte poskytnutý súbor CSV a navrhnite, ktorých položiek by mal obchod nakupovať viac/menej. Odôvodnite svoje zdôvodnenie.

```python
import pandas as pd

df = pd.read_csv("csvsample.csv")
quantity_purchase = df[df["Type"] == "Purchase"]
quantity_sale = df[df["Type"] == "Sale"]

total_purchase_quantity = quantity_purchase.groupby("Item")["Quantity"].sum().rename("Quantity_purchased")
total_sale_quantity = quantity_sale.groupby("Item")["Quantity"].sum().rename("Quantity_sold")
summary = pd.merge(total_purchase_quantity, total_sale_quantity, on="Item", how="outer").fillna(0)

summary["Difference"] = summary["Quantity_purchased"] - summary["Quantity_sold"]

summary["Recomendation"] = summary["Difference"].apply(
    lambda x: "Buy more" if x < 0 else ("Buy less" if x > 0 else "OK")
)

print(summary)
```

## ⭐ Úloha 10
Prečítajte textový súbor a vytlačte len riadky, ktoré sú dlhšie ako 30 znakov.

```python
with open("textsample.txt", "r", encoding="utf-8") as file:
    content = file.read()
    lines = content.split("\n")
    
    for line in lines:
        character_count = len(line)
        if character_count > 30:
        
            print(f"Riadok: '{line}' ma {character_count} znakov.")
```

## ⭐ Úloha 11
Prečítajte textový súbor a nájdite najdlhší riadok. Vytlačte číslo riadku, obsah riadku a jeho dĺžku (počet znakov).

```python
with open("textsample.txt", "r", encoding="utf-8") as file:
    content = file.read()
    lines = content.split("\n")
    
    the_longest_line = " "
    max_length = 0
    line_number = 0
    
    for i, line in enumerate(lines, start=1):
        character_count = len(line)
        if character_count > max_length:
            max_length = character_count
            the_longest_line = line
            line_number = i 
    
    print(f"The longest line is number {line_number}:")
    print(f"{the_longest_line}")
    print(f"It has {max_length} characters.")
    
```
