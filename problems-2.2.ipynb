# Úloha 01

Napíš funkciu safe_divide, ktorá berie dve čísla a a b a vydelí tieto dve čísla. Uisti sa, že vyhodíš príslušné exceptions, keď a a b majú nevhodné hodnoty.

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print ("Error: Cannot divide by zero!")
        result = None
    except TypeError:
        print("Error: Both a and b must be numbers.")
        result = None
    finally:
        print("Division operation completed.")
    return result   
```

```python
print(safe_divide(10, 5))

print(safe_divide(8, 13))

print(safe_divide(7, 0))

print(safe_divide(8, "a"))
```

# Úloha 02

Napíš funkciu check_date, ktorá berie tri stringy day, month a year a skontroluje, že poskytnuté hodnoty zodpovedajú legitímnemu dátumu narodenia.

```python
from datetime import date

def check_date(day, month, year):
    try:
        check_date = date(year, month, day)
        return True, check_date
    except ValueError:
        return False, None
```

```python
print(check_date(31, 12, 2025))
print(check_date(30, 2, 2024))
print(check_date(31, 6, 2025))
print(check_date(30, 7, 2025))
```

# Úloha 03

Pracuješ pre obchod, ktorý ponúka akciu "kúp 3, získaj 1 zadarmo" na jablká. Momentálne je v obchode 400 jabĺk. Napíš funkciu buy_apples, ktorá berie počet jabĺk, ktoré chce zákazník kúpiť, a vráti konečnú cenu košíka.

```python
import math 

def buy_apples(apples):
    max_amount = 400
    price = 1
    if apples > max_amount:
        return "Nie je dostatok jablk v obchode"
    
    skupiny = math.floor(apples / 4)
    zostatok = apples % 4
    
    total_apples = skupiny * 3 + zostatok
    total_prices = total_apples * price
    
    return f"Zakaznik kupi {apples} jablk a zaplati {total_prices: } EUR."

print(buy_apples(8))
print(buy_apples(7))
print(buy_apples(407))
    
```

# Úloha 04

Napíš funkciu calculate_parking_fee, ktorá berie duration parkovania v minútach a vráti cenu, s rôznymi sadzbami pre rôzne časové obdobia (napr. 0-15 minút je účtovaných inou sadzbou ako 1-2 hodiny).

```python
import math

def calculate_parking_fee(duration_minutes):
    if duration_minutes <= 15:
        return 0
    elif duration_minutes <= 60:
        return 0.6
    elif duration_minutes <= 120:
        return 4
    else:
        extra_minutes = duration_minutes - 120
        extra_hours = math.ceil(extra_minutes / 60)
        extra_fee = extra_hours * 1
        return 4 + extra_fee
    
print(calculate_parking_fee(150))
print(calculate_parking_fee(40))
print(calculate_parking_fee(240))
            
```

Napíš funkciu withdraw_money, ktorá berie stav účtu a sumu výberu, a vráti nový stav účtu po validácii transakcie.

```python
import math
from decimal import Decimal, ROUND_DOWN

def withdraw_money(stav_uctu, vyber):
    if not math.isfinite(stav_uctu) or not math.isfinite(vyber):
        return "Neplatny vstup - musi byt cislo"
    
    stav_uctu = Decimal(str(stav_uctu))
    vyber = Decimal(str(vyber))
    
    if vyber <= 0:
        return "Neplatna suma na vyber"
    if vyber > stav_uctu:
        return "Nedostatok penai na ucte"
    
    novy_stav_uctu = (stav_uctu - vyber).quantize(Decimal("0.01"), rounding=ROUND_DOWN)
    
    return f"Transakcia uspesna. Novy zostatok: {novy_stav_uctu} EUR"

print(withdraw_money(125.50, 36.26))
    
```

Napíš funkciu validate_credit_card, ktorá berie string s číslom kreditnej karty a určí, či je validné pomocou Luhn algoritmu. Otestuj na "4532015112830366" a "1234567890123456".

```python
def validate_credit_card(credit_card_number):
    rev_credit_card_number = [int(n) for n in credit_card_number[::-1]]
    doubled_list = 0
    
    for n in range(1, len(rev_credit_card_number), 2):
        doubled = rev_credit_card_number[n] * 2
        if doubled > 9:
           doubled -=9
        doubled_list += doubled
            
    result = doubled_list % 10 == 0
    return result
      
   
credit_card_number = "4532015112830366"
    
print(validate_credit_card(credit_card_number))
```

# Úloha 07

Vytváraš booking systém pre kino. Napíš funkciu book_seats, ktorá berie list čísel sedadiel (vo formáte trojznakového stringu pozostávajúceho z jedného písmena a dvoch čísel) a zabezpečí, že sú dostupné a validné.

```python
import re

rows = ["A", "B", "C", "D", "E", "F"]
columns = [f"{i:02}" for i in range(1, 21)]

booked = {"A03", "C11", "F08"}
        
def book_seats(seats):
    global booked
    new_bookings = []
        
    for seat in seats:
        if not re.match(r"^[A-F]\d{2}$", seat):
            print(f"Sedadlo {seat} ma nespravny format.")
            continue
        
        row = seat[0].upper()
        num = seat[1:]
        
        if row not in rows or num not in columns:
            print(f"Sedadlo {seat} je mimo rozsahu.")
            continue
        
        if seat in booked:
            print(f"Sedadlo {seat} je uz obsadene.")
            continue
        
        booked.add(seat)
        new_bookings.append(seat)
             
    
    if new_bookings:
        print("Uspesne rezervovane sedadla:", ", ".join(new_bookings))
  
book_seats(["A01", "A03", "C11", "E05", "Z10"])
book_seats(["A07", "C12", "D03", "X58", "F24", "123", "E13"])
```

# Úloha 08

Napíš funkciu generate_username, ktorá berie krstné meno, priezvisko, rok narodenia a databázu a vráti tri random možné usernames, pričom zabráni duplikátom v databáze.

```python
import random
import string

def generate_username(first_name, last_name, year, database):
    suggestions = set() 
    
    while len(suggestions) < 3:
        base = (first_name[0] + last_name). lower()
        
        option = random.choice([
            base + str(year)[-2],
            base + str(random.randint(10, 99)),
            base + random.choice(string.ascii_lowercase),
            first_name.lower() + last_name[0].lower() + str(random.randint(1, 999))
        ])
        
        if option not in database:
            suggestions.add(option)

    return list(suggestions)    

print(generate_username("julia", "bagdar", 1995, database=set()))
    
```

# Úloha 09

Napíš funkciu find_mutual_friends, ktorá berie user ID a vráti všetkých userov, ktorí sú kamaráti aspoň s 3 kamarátmi daného usera. Dokáž, že tvoj algoritmus funguje, keď má (a nefunguje, keď nemá).

```python
from collections import Counter

def find_mutual_friends(user_id, friends):

    user_friends = set(friends.get(user_id, []))
    mutual_count = Counter()
    
    for friend in user_friends:
        for fof in set(friends.get(friend, [])):
            if fof != user_id and fof not in user_friends:
                mutual_count[fof] += 1

    result = []
    
    for u, count in mutual_count.items():
         if count >= 3:
             result.append(u)
    
    return result

friends = {
    "A": ["B", "C", "D", "E", "F"],
    "B": ["A", "C", "D", "G"],       # B má kamarátov C, D, G
    "C": ["A", "B", "D", "G"],       # C má kamarátov B, D, G
    "D": ["A", "B", "C", "G"],       # D má kamarátov B, C, G
    "E": ["A", "F"],
    "F": ["A", "E"],
    "G": ["B", "C", "D"],             # G má teraz 3 spoločných kamarátov s A
}


print(find_mutual_friends("A", friends))
print(find_mutual_friends("J", friends))
print(find_mutual_friends("C", friends))
```

# Úloha 10

Napíš funkciu calculate_shipping_cost, ktorá berie hmotnosť balíka, rozmery a PSČ destinácie, a vráti cenu doručenia. Zohľadni, že rôzne parametre sa účtujú rôznymi sadzbami.

```python
import math

def calculate_shipping_cost(weight, length, width, height, postal_code):
    base_cost = 5.0
    weight_cost = weight * 1
    volume = length * width * height
    volume_cost = volume * 0.001
    
    cost = base_cost + weight_cost + volume_cost
    
    if str(postal_code).startswith("1"):
        cost *= 1.10
        
    cost = math.ceil(cost)
    return cost


price = calculate_shipping_cost(1.2, 30, 20, 10, "04001")
print("Cena dorucenia:", price, "EUR")

price = calculate_shipping_cost(5, 40, 25, 15, "05801")
print("Cena dorucenia:", price, "EUR")
```
