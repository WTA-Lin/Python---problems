## ⭐ Úloha 01

Napíš funkciu, ktorá berie ľubovoľný string a vráti počet číslic v tomto stringu.

```python
import re

def counter(text): 
    pocet = 0
    for ch in text:
        if ch.isdigit():
            pocet += 1   
    return pocet

## string = "Kompas preskočil o 30°, a Alex vedel, že je to znamenie — poklad ukrytý za kameňmi s číslami 12 – 7 – 3 už nie je ďaleko."
       
zoznam = ["lldjoifjeojo254e", "lkodiofeiae", "254d5joeijoei"]

for string in zoznam:
    counter(string)
    
    print(counter(string))
```

## ⭐ Úloha 02

Napíš funkciu, ktorá vezme string a vráti ten istý string, ale písmená budú na striedačku malé a veľké (tAkTo).

```python
def velke_male(trenuj):   
    upraveny_text = ""

    for pismenko in range(len(trenuj)):
        if pismenko % 2 == 0:
            upraveny_text += trenuj[pismenko].lower()
        else:
            upraveny_text += trenuj[pismenko].upper()  
    return upraveny_text
        
text = "Kompas preskočil o 30°, a Alex vedel, že je to znamenie — poklad ukrytý za kameňmi s číslami 12 – 7 – 3 už nie je ďaleko."

print(velke_male(text))
```
