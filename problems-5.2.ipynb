## ⭐ Úloha 03

Zoraď dáta v tomto liste pomocou ktoréhokoľvek triediaceho algoritmu.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

```python
sorted_nums = merge_sort(alkohol)
print(sorted_nums)
```

## ⭐ Úloha 04

Bez toho, aby si sa pozrela na svoj list dát, vytvor list s tromi hodnotami, ktoré by mohli byť v tvojich dátat. Potom implementuj hociktorý vyhľadávací algoritmus a nájdi index týchto hodnôt, ak existuje.

```python
sorted_nums = merge_sort(alkohol)

target_list = [8.4, 9.0, 15.0]

for target in target_list:
    left = 0
    right = len(sorted_nums) - 1
    found_index = -1

    while left <= right:
        mid = (left + right) // 2
            
        if sorted_nums[mid] == target:
            found_index = mid
            break
        elif sorted_nums[mid] < target:
            left = mid + 1 
        else:
            right = mid - 1 
        
    print(f"Hladam {target} na indexe {found_index}")
```
