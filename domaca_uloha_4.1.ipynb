Použi funkciu groupby na nájdenie celkového počtu predaných kusov na daný produkt zaznamenaných v csvsample.csv.

```python
import pandas as pd
df = pd.read_csv("csvsample.csv")
sales = df[df["Type"] == "Sale"]
total_sale = sales.groupby("Item")["Quantity"].sum()
total_sale
```
